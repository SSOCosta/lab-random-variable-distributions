# Lab | Random variable distributions

```python
# Refer to the numerical.csv dataset that you previously saved.
# 1.Use the numerical variables from our dataset this week.
import pandas as pd
import numpy as np

numerical_df = pd.read_csv('numerical.csv')
numerical_df.head()
```

```python
# 2.Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
import seaborn as sns
import matplotlib.pyplot as plt

for column in numerical_df.columns:
    custom_palette = sns.color_palette("Set2")
    sns.set_palette(custom_palette)
    plt.figure(figsize=(8, 6))
    sns.histplot(numerical_df[column], kde=True)  # You can use 'kde=False' to hide the kernel density estimate
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    
'''None of the plots has normal distribution, they are right skewed with tail do the right. Months_since_policy_inception is 
the only one that differs from all the others and perhaps what could look more similar to a normal distribution shape.
They are all positively skewed.'''
```

```python
# 3.Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

# Creating a copy of the DataFrame to store the log-transformed data
log_transformed_df = numerical_df.copy()

# Applying the log transformation to each column
for column in numerical_df.columns:
        log_transformed_df[column] = np.log(numerical_df[column] + 1) 

# Plotting histograms for each transformed column
for column in log_transformed_df.columns:
        plt.figure(figsize=(8, 4))
        sns.histplot(log_transformed_df[column], kde=True, color="blue")  
        plt.title(f'Distribution of Log-Transformed {column}')
        plt.xlabel(f'Log({column} + 1)')
        plt.ylabel('Frequency')
        plt.show()

'''After this log transformation we can observe that the graph of 'customer_lifetime_value' has now a more similar shape
to a normal distribution.
income became left skewed once before was right skewed.
montlhy_premium_auto remained right skewed, however, with a much smaller tail to the right.
months_since_last_claim was converted sides of skew, from right to left, but overal remais de same
months_since_policy_inception suffered the major transformation having now all the values skewed to the left.
number_of_open_complaints and number_of_policies look to remais the same.
total_claim_amount change from a right skew to left skew, however, with a higher aproximation to shape of a normal distribution
apart of the long left tail. 
Because we applied the log function, most of graphs with a positive skew became negative skewed.
We can also observe less long tails, due to the "power" of log function to bring(scalling) them to a nearest normal distribution position'''
```
